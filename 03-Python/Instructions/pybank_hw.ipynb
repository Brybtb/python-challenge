```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Pybank/Resources/budget_data.csv", parse_dates=['Date'], dtype={'Profit/Losses': float})
df.sort_values(by='Date')
```

```python
# of months
df['Date'].dt.month.count()
```

```python
# Total net Profits/Losses
df['Profit/Losses'].sum()
```

```python
# Average change in Profit/Losses between months
df['Prior P&L'] = df['Profit/Losses'].shift(1)#.bfill()
df['Change'] = df['Profit/Losses'] -df['Prior P&L'] 
df.Change.mean()
```

```python
df['Prior P&L'] = df['Profit/Losses'].shift(1)#.bfill()
df['Change'] = df['Profit/Losses'] - df['Prior P&L'] 
print(df['Change'].sum())
print(df['Change'].count())
print(df['Change'].sum() / df['Change'].count())
```

```python
# Greatest increase in profits over the entire period
df.loc[df.Change == df.Change.max()]
```

```python
max_change = df.Change.max()
df.loc[df.Change == max_change]
new_df = df[df.Change == max_change]
new_df
```

```python
min_change = df.Change.min()
df.loc[df.Change == min_change]
new_df1 = df[df.Change == min_change]
new_df1
```
